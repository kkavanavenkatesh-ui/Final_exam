```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Loading datasets...")


trader_df = pd.read_csv('historical_data.csv', low_memory=False)
sentiment_df = pd.read_csv('fear_greed_index.csv')

print("Processing timestamps and engineering date keys...")


sentiment_df['DateKey'] = pd.to_datetime(sentiment_df['date'], errors='coerce').dt.date


if pd.api.types.is_numeric_dtype(trader_df['Timestamp']):
    trader_df['DateKey'] = pd.to_datetime(trader_df['Timestamp'], unit='ms').dt.date
else:
    trader_df['DateKey'] = pd.to_datetime(trader_df['Timestamp'], errors='coerce').dt.date



trader_df = trader_df.dropna(subset=['DateKey'])
sentiment_df = sentiment_df.dropna(subset=['DateKey'])

print("Merging trader dataset with Bitcoin Fear & Greed sentiment index...")


df = pd.merge(trader_df, sentiment_df, on='DateKey', how='inner')
print(f"Successfully matched and merged {len(df)} trade records!")

df['is_win'] = df['Closed PnL'] > 0

print("\n" + "="*50)
print("   1. PERFORMANCE & RISK METRICS BY SENTIMENT")
print("="*50)

summary = df.groupby('classification').agg(
    Total_Trades=('Account', 'count'),
    Win_Rate=('is_win', 'mean'),
    Total_PnL=('Closed PnL', 'sum'),
    Avg_PnL=('Closed PnL', 'mean'),
    Median_PnL=('Closed PnL', 'median'),
    Avg_Trade_Size_USD=('Size USD', 'mean')
).reset_index()

summary['Win_Rate'] = (summary['Win_Rate'] * 100).round(2)
summary['Total_PnL'] = summary['Total_PnL'].round(2)
summary['Avg_PnL'] = summary['Avg_PnL'].round(2)
summary['Median_PnL'] = summary['Median_PnL'].round(2)
summary['Avg_Trade_Size_USD'] = summary['Avg_Trade_Size_USD'].round(2)

print(summary.to_string(index=False))

print("\n" + "="*50)
print("   2. DIRECTIONAL BIAS (LONG VS SHORT BEHAVIOR)")
print("="*50)

direction_bias = pd.crosstab(df['classification'], df['Side'], normalize='index') * 100
print(direction_bias.round(2))

print("\nGenerating visual performance insights...")
sns.set_theme(style="darkgrid")

plt.figure(figsize=(10, 5))
sns.barplot(x='classification', y='Win_Rate', data=summary, palette='viridis')
plt.title('Hyperliquid Trader Win Rate (%) Across Market Sentiment Phases')
plt.xlabel('Fear & Greed Index Classification')
plt.ylabel('Win Rate %')
plt.tight_layout()
plt.savefig('sentiment_vs_winrate.png')

print("Analysis completely done! Visual plots exported as 'sentiment_vs_winrate.png'.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Panel 1: Clean Win Rate Comparison
sns.barplot(x='classification', y='Win_Rate', hue='classification', data=summary, palette='viridis', legend=False, ax=axes[0])
axes[0].set_title('Trader Win Rate (%) Across Sentiment Phases', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Win Rate %')
axes[0].set_xlabel('Market Sentiment')

# Panel 2: Average Profit/Loss per Trade Breakdown
sns.barplot(x='classification', y='Avg_PnL', hue='classification', data=summary, palette='magma', legend=False, ax=axes[1])
axes[1].set_title('Average Profit per Trade ($) Across Sentiment Phases', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Average PnL ($)')
axes[1].set_xlabel('Market Sentiment')

plt.suptitle('Hyperliquid Multi-Dataset Deep Dive: Sentiment vs Performance Analytics', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('hyperliquid_sentiment_analytics_dashboard.png', bbox_inches='tight')
print("Dashboard visualization successfully saved as 'hyperliquid_sentiment_analytics_dashboard.png'!")
```
